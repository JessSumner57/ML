```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

```

```python
# csv
df = pd.read_csv('Bike-Sharing-Dataset/hour.csv')


```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['cnt', 'dteday'], axis=1), df['cnt'], test_size=0.2)

model = RandomForestRegressor()
model.fit(X_train, y_train)

score = model.score(X_test, y_test)
print('R-squared score:', score)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


#avg number of bikes rented
hourly_data = df.groupby('hr').agg({'cnt': 'mean'})

#creating a line plot to look at avg number of bikes rented by hour
plt.plot(hourly_data.index, hourly_data['cnt'])
plt.title('Average Number of Bikes Rented by Hour')
plt.xlabel('Hour')
plt.ylabel('Number of Bikes Rented')
plt.show()
```

```python
import pandas as pd
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop(['cnt', 'dteday'], axis=1), df['cnt'], test_size=0.2)

# Train a neural network regression model on the training data
model = MLPRegressor(hidden_layer_sizes=(100, 50))
model.fit(X_train, y_train)

# Evaluate the model on the testing data
y_pred = model.predict(X_test)
score = r2_score(y_test, y_pred)
print('R-squared score:', score)
```

```python
from sklearn.svm import SVR



# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop(['cnt', 'dteday'], axis=1), df['cnt'], test_size=0.2)

# Train a support vector regression model on the training data
model = SVR()
model.fit(X_train, y_train)

# Evaluate the model on the testing data
y_pred = model.predict(X_test)
score = r2_score(y_test, y_pred)
print('R-squared score:', score)
```
